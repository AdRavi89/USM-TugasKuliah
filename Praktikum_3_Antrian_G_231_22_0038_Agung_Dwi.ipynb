# **ANTRIAN dengan Tambahan Tampilkan isi Object Tersimpan**

```python
import os
import queue

class myQueue:
    def __init__(self):
        self.items = queue.Queue()
        self.item_order = 1  # Menambahkan variabel untuk melacak urutan objek

    # Memeriksa apakah queue dalam keadaan kosong
    def isEmpty(self):
        return self.items.empty()

    # Menambah data ke queue
    def qPut(self, item):
        self.items.put((self.item_order, item))  # Menambahkan nomor indeks ke setiap objek
        self.item_order += 1

    # Mengeluarkan data dari queue
    def qGet(self):
        if not self.items.empty():
            return self.items.get()[1]  # Mengambil objek saja
        else:
            return "empty"

    # Menghitung panjang Queue
    def size(self):
        return self.items.qsize()

    # Main menu aplikasi
    def mainmenu(self):
        pilih = "y"
        while (pilih == "y"):
            os.system("clear")

            print("=" * 20)
            print("Menu Aplikasi Queue")
            print("=" * 20)
            print("1. Input Object")
            print("2. Get Object")
            print("3. Cek Empty")
            print("4. Panjang dari Queue")
            print("5. Tampilkan Isi Queue")
            print("=" * 20)
            pilihan = str(input("Silahkan masukan pilihan anda: "))
            print("=" * 20)
            if (pilihan == "1"):
                os.system("clear")
                obj = str(input("Masukan objek yang ingin anda tambahkan: "))
                self.qPut(obj)
                print("Object " + obj + " telah ditambahkan")
                x = input("")

            elif (pilihan == "2"):
                os.system("clear")
                temp = self.qGet()
                if temp != "empty":
                    print("Object " + temp + " telah Dihapus")
                else:
                    print("Queue Kosong")
                x = input("")
            elif (pilihan == "3"):
                os.system("clear")
                print(self.isEmpty())
                x = input("")
            elif (pilihan == "4"):
                os.system("clear")
                print("Panjang dari Queue adalah : " + str(self.size()))
                x = input("")
            elif (pilihan == "5"):
                os.system("clear")
                self.showQueue()
                x = input("")
            else:
                pilih = "n"

    # Menampilkan isi dari Queue dengan urutan
    def showQueue(self):
        if self.isEmpty():
            print("Queue Kosong")
        else:
            print("Isi Queue:")
            items = list(self.items.queue)
            for item in items:
                order, obj = item
                print(f"{order}. {obj}")

if __name__ == "__main__":
    q = myQueue()
    q.mainmenu()
```
