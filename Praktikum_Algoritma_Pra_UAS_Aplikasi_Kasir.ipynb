Aplikasi KASIR

Input :

Nomor Transaksi, Nama Barang, Harga, Jumlah Beli, Member (Yes/No)

Proses :

Jika Jumlah Beli:

Jumlah Beli > 5 = Diskon 5%

Jumlah Beli > 10 = Diskon 7%

Jumlah Beli > 15 = Diskon 10%

Jumlah Beli > 25 = Diskon 15%

Jika Member = Diskon Member 5%

Total bayar = ((Harga * Jumlah beli) - Diskon) - Diskon Member



Output :

Nomor Transaksi :

Nama Barang:

Harga :

Jumlah Beli :

Total :

Diskon :

Diskon Member :

Total Bayar  :



Buatlah laporan dalam format PDF (Script + hasil output)

```python
#PRAUAS - A2

print("---------------APLIKASI KASIR TOKO---------------")
no_transaksi=int(input("Nomor Transaksi : "))
nama_barang=(input("Nama Barang     : "))
harga=int(input("Harga           : Rp."))
jumlah_beli=int(input("Jumlah Beli     : "))
member=(input("Member (Y / N)  : "))
print("----------------------------------------------")
total = harga * jumlah_beli

if jumlah_beli > 25 :
  diskon_barang = total * (15/100)
elif jumlah_beli > 15 :
  diskon_barang = total * (10/100)
elif jumlah_beli > 10 :
  diskon_barang = total * (7/100)
elif jumlah_beli > 5 :
  diskon_barang = total * (5/100)
else :
  diskon_barang = 0

if member =="y" :
    diskon_member = total*0.05
else :
  diskon_member = 0

total_bayar = ((harga*jumlah_beli) - diskon_barang) - diskon_member
print("---------------------------------------------")
print("---------------------------------------------")
print("no_transaksi     : ", no_transaksi)
print("nama_barang      : ", nama_barang)
print("harga            : Rp {:,}".format(int(harga)).replace(',','.'))
print("jumlah_beli      : ", jumlah_beli)
print("---------------------------------------------")
print("---------------------------------------------")
print("total             : Rp {:,}".format(int(total)).replace(',','.'))
print("---------------------------------------------")
print("diskon_barang     : Rp {:,}".format(int(diskon_barang)).replace(',','.'))
print("dikon_member      : Rp {:,}".format(int(diskon_member)).replace(',','.'))
print("---------------------------------------------")
print("total_bayar       : Rp {:,}".format(int(total_bayar)).replace(',','.'))
print("---------------------------------------------")
print("---------------------------------------------")
```

Aplikasi Sewa Mobil

```python

print("********************************************")
print("--------Aplikasi Sewa Mobil.")
print("--------Jenis Mobil : ")
print("--------Avanza, Inova, Pajero, Minicoper, dan Tesla")
print("********************************************")
avanza=250000
inova=350000
pajero=500000
minocoper=1000000
tesla=2000000

print("---------------------------------------------")
print("---------------------------------------------")
nomerTransaksi=int(input("Nomer Transaksi       = "))
print("---------------------------------------------")
namaPenyewa=input("Nama Penyewa          = ")
alamat=input("Alamat                = ")
print("---------------------------------------------")
print("---------------------------------------------")
lamaSewa=int(input("Lama Sewa Berapa Hari = "))
mobil=input("Mobil                 = ")
print("---------------------------------------------")
print("---------------------------------------------")
if mobil == "avanza":
  harga=avanza*lamaSewa
  if lamaSewa > 3:
      diskon=avanza*0.1

      print("Harga/Hari            = Rp {:,}".format(int(avanza)).replace(',','.'))
      print("Diskon Sewa > 3 hari  = 10%")
  else :
      diskon = harga
      print("Diskon = anda tidak mendapat diskon")

elif mobil == "inova":
  harga=inova*lamaSewa
  if lamaSewa > 3:
      diskon=inova*0.1

      print("Harga/Hari            = Rp {:,}".format(int(inova)).replace(',','.'))
      print("Diskon Sewa > 3 hari  = 10%")
  else :
      diskon = harga
      print("Diskon = anda tidak mendapat diskon")

elif mobil == "pajero":
  harga=pajero*lamaSewa
  if lamaSewa > 3:
      diskon=pajero*0.1

      print("Harga/Hari            = Rp {:,}".format(int(pajero)).replace(',','.'))
      print("Diskon Sewa > 3 hari  = 10%")
  else :
      diskon = harga
      print("Diskon = anda tidak mendapat diskon")

elif mobil == "minicoper":
  harga=minocoper*lamaSewa
  if lamaSewa > 3:
      diskon=minicoper*0.1

      print("Harga/Hari            = Rp {:,}".format(int(minicoper)).replace(',','.'))
      print("Diskon Sewa > 3 hari  = 10%")
  else :
      diskon = harga
      print("Diskon = anda tidak mendapat diskon")

elif mobil == "tesla":
  harga=tesla*lamaSewa
  if lamaSewa > 3:
      diskon=tesla*0.1

      print("Harga/Hari            = Rp {:,}".format(int(tesla)).replace(',','.'))
      print("Diskon Sewa > 3 hari  = 10%")
  else :
      diskon = harga
      print("Diskon                = Anda tidak mendapat diskon")
else:
  print("mohon masukan data yang benar")

biayasewa=harga-diskon
print("---------------------------------------------")
print("Total Biaya Sewa      = Rp {:,}".format(int(biayasewa)).replace(',','.'))
print("---------------------------------------------")
print("---------------------------------------------")
```
