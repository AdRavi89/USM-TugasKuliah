```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def print_postorder(self, node):
        if node is not None:
            self.print_postorder(node.left)
            self.print_postorder(node.right)
            print(node.key, end=" ")

    def print_inorder(self, node):
        if node is not None:
            self.print_inorder(node.left)
            print(node.key, end=" ")
            self.print_inorder(node.right)

    def print_preorder(self, node):
        if node is not None:
            print(node.key, end=" ")
            self.print_preorder(node.left)
            self.print_preorder(node.right)

    def print_postorder_tree(self):
        self.print_postorder(self.root)

    def print_inorder_tree(self):
        self.print_inorder(self.root)

    def print_preorder_tree(self):
        self.print_preorder(self.root)

    def build_tree(self):
        if self.root is not None:
            print("Binary tree already exists. Cannot build again.")
            return

        key_list = input("Enter the list of keys separated by space: ").split()
        self.root = self.build_tree_recursive(key_list)

    def build_tree_recursive(self, key_list):
        if not key_list:
            return None

        key = key_list.pop(0)
        if key == "None":
            return None

        node = Node(int(key))
        node.left = self.build_tree_recursive(key_list)
        node.right = self.build_tree_recursive(key_list)

        return node


class SequentialSearch:
    def __init__(self):
        self.all_data = []

    def display_data(self):
        if not self.all_data:
            print("Array is empty.")
        else:
            for data in self.all_data:
                print(data, end=" ")
            print()

    def searching(self, karakter):
        x = 0
        ketemu = False
        for i in range(x, len(self.all_data)):
            if karakter == self.all_data[i]:
                ketemu = True
                x = i
        if ketemu:
            print("Data berada pada urutan ke -", x + 1)
        else:
            print("Data Tidak Ditemukan")

    def input_data(self):
        n = int(input("Enter the number of elements in the array: "))
        for i in range(n):
            element = input(f"Enter element {i + 1}: ")
            self.all_data.append(element)


if __name__ == "__main__":
    binary_tree = BinaryTree()
    sequential_search = SequentialSearch()

    while True:
        print("\nMenu:")
        print("1. Input data for binary tree")
        print("2. Input data for array")
        print("3. Display binary tree (preorder, inorder, postorder)")
        print("4. Display array")
        print("5. Search in array")
        print("6. Exit")

        choice = int(input("Enter your choice (1-6): "))

        if choice == 1:
            binary_tree.build_tree()
        elif choice == 2:
            sequential_search.input_data()
        elif choice == 3:
            print("\nPreorder traversal of binary tree:")
            binary_tree.print_preorder_tree()
            print("\nInorder traversal of binary tree:")
            binary_tree.print_inorder_tree()
            print("\nPostorder traversal of binary tree:")
            binary_tree.print_postorder_tree()
        elif choice == 4:
            print("\nArray:")
            sequential_search.display_data()
        elif choice == 5:
            search_key = input("Enter the element to search in the array: ")
            sequential_search.searching(search_key)
        elif choice == 6:
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")
```
